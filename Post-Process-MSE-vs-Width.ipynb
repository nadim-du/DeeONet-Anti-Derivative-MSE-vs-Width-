```python
import os 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
d2_train = np.load('2A-Data/d2_train.npy')
d2_test = np.load('2A-Data/d2_test.npy')

d3_train = np.load('2A-Data/d3_train.npy')
d3_test = np.load('2A-Data/d3_test.npy')

d4_train = np.load('2A-Data/d4_train.npy')
d4_test = np.load('2A-Data/d4_test.npy')
```

```python
widths = np.arange(10, 2560, 100)

fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(widths, d2_train, marker = 'o', color = 'blue', label = 'Depth 2:Train')    
ax.plot(widths, d2_test, marker = 'o', linestyle = '--', color = 'blue', label = 'Depth 2:Test')

ax.plot(widths, d3_train, marker = 's', color = 'red', label = 'Depth 3:Train')    
ax.plot(widths, d3_test, marker = 's', linestyle = '--', color = 'red', label = 'Depth 3: Test')  

# ax.plot(widths, d4_train, marker = 'd', color = 'green', label = 'Depth 4:Train')    
# ax.plot(widths, d4_test, marker = 'd', linestyle = '--', color = 'green', label = 'Depth 4:Test')


plt.xlim([0, 900])
plt.ylim([0, 0.2])

plt.legend()


plt.xlabel('Width')
plt.ylabel('MSE')

plt.savefig('MSE Vs Width.jpg', dpi = 500)
```
